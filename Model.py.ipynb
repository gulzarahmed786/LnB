```python
from ultralytics import YOLO
from collections import Counter
# Load the trained YOLOv8n model
#with your custom weight file
model = YOLO("D:/best.pt")
# Get the class names from the
class_names = model.names
# Run inference on a list of new
#image_paths = ['10.jpeg', '4.jpg']
results = model(r"C:\Users\gzar9\Downloads\WhatsApp Image 2024-09-14 at 7.28.42 PM.jpeg")
```

```python
results
```

```python
detected_objects = results[0].boxes
num_objects = len(detected_objects)
class_ids = [box.cls.item() for box in detected_objects]
class_counts = Counter(class_ids)

class_names = ['Bonnet', 'Bumper', 'Dickey', 'Door', 'Fender', 'Light', 'Windshield']

# Display the counts for each class
for class_id, count in class_counts.items():
    class_name = class_names[int(class_id)]  # Get the class name using the ID
    print(f"{class_name} Count: {count}")
```
